```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("insurance.csv")
```

```python
X = df.drop('charges', axis=1)
y = df['charges']
```

```python
cat_cols = ['sex', 'smoker', 'region']
num_cols = ['age', 'bmi', 'children']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
preprocessor = ColumnTransformer([('encoder', OneHotEncoder(drop='first'), cat_cols)], remainder='passthrough')
```

```python
X_processed = preprocessor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_processed , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
print(r2_score(y_test , y_pred))
```

```python
scores = []
for i in range (1000):
    X_train , X_test , y_train , y_test = train_test_split(X_processed , y , test_size = 0.2 , random_state = i)
    lr = LinearRegression()
    lr.fit(X_train , y_train)
    y_pred = lr.predict(X_test)
    scores.append(r2_score(y_test , y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_processed , y , test_size = 0.2 , random_state = np.argmax(scores))
lr = LinearRegression()
lr.fit(X_train , y_train)
y_pred = lr.predict(X_test)
print(r2_score(y_test , y_pred))
```

```python
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump((lr, preprocessor), file)
```
